```python
results['avito_values'] = (
    results['avito_values']
    .str.lower()
    .str.replace("'", "") # заодно уберем апострофы 
    .str.replace(",", "") # и запятые
)
```

```python
# создаем список с ценностями
check = ['мы ставим потребности пользователей на первое место',
         "put users needs first", 
         'мы доверяем своей команде', 
         'in team we trust',
         'результат начинается с тебя',  
         'results start with you', 
         'мы решаем проблемы через коммуникацию', 
         'solve problems through communications', 
         'мы всегда стремимся к большему', 
         'strive to improve', 
         'мы принимаем решения основанные на данных', 
         'use data to decide', 
         'упрощай', 
         'keep it simple']

raw_data = {'values': check} # создаем словарь
df = pd.DataFrame(raw_data, columns = ['values']) # создаем новый датафрейм

counts = [] 
for i in check:
    inputs = results['avito_values'].str.count(i).sum() # считаем сумму вхождений по столбцу
    counts.append(inputs)
       
df['counts'] = counts
df
```

```python
# сделаем замены, посмотрим на результаты на русском языке

df['values'] = (
    df['values']
    .str.replace('put users needs first',
                 'мы ставим потребности пользователей на первое место')
    .str.replace('in team we trust',
                 'мы доверяем своей команде')
    .str.replace('results start with you',
                 'результат начинается с тебя')
    .str.replace('solve problems through communications',
                 'мы решаем проблемы через коммуникацию')
    .str.replace('strive to improve',
                 'мы всегда стремимся к большему')
    .str.replace('use data to decide',
                 'мы принимаем решения основанные на данных')
    .str.replace('keep it simple',
                 'упрощай')
)    
```
